```python
#make it 
soup = BeautifulSoup(browser.page_source,"lxml")
```

```python
results = soup.find_all('div', {'class' : 'vehicle-card'})
```

```python
nama = []
jarak_tempuh = []
nama_dealer = []
rate = []
banyak_review = []
harga = []

for result in results:
    
    # untuk mencari nama
    try:
        nama.append(result.find('h2').get_text()) 
    except:
        nama.append('n/a')
    
    # untuk mencari jarak tempuh
    try:
        jarak_tempuh.append(result.find('div', {'class':'mileage'}).get_text())
    except:
        jarak_tempuh.append('n/a')
    
    # untuk mencari nama dealer
    try:
        nama_dealer.append(result.find('div', {'class':'dealer-name'}).get_text().strip())
    except:
        nama_dealer.append('n/a')
    # untuk mencari rate
    try:
        rate.append(result.find('span', {'class':'sds-rating__count'}).get_text())
    except:
        rate.append('n/a')
    
    # untuk mencari banyak review
    try:
        banyak_review.append(result.find('span', {'class':'sds-rating__link'}).get_text())
    except:
        banyak_review.append('n/a')
        
    #untuk mencari harga 
    try:
        harga.append(result.find('span', {'class':'primary-price'}).get_text())
    except:
        harga.append('n/a')    
    
```

```python
dealer_mobil = pd.DataFrame({'nama': nama, 'jarak tempuh': jarak_tempuh, 'nama dealer': nama_dealer,'rate': rate, 'banyak review': banyak_review, 'harga': harga})
```

```python
#untuk data cleaning 
dealer_mobil['banyak review'] = dealer_mobil['banyak review'].apply(lambda x: x.strip('review)').strip('('))
```

```python
#Part 2 bagian halaman lebih dari 1

nama = []
jarak_tempuh = []
nama_dealer = []
rate = []
banyak_review = []
harga = []

for i in range (1,70):
    
    # website in variable
    website = 'https://www.cars.com/shopping/results/?page='+ str(i) +'&page_size=20&dealer_id=&list_price_max=&list_price_min=&makes[]=mercedes_benz&maximum_distance=20&mileage_max=&sort=best_match_desc&stock_type=cpo&year_max=&year_min=&zip=' 
    
    # request to website
    response = requests.get(website)
    
    # soup object
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # results
    results = soup.find_all('div', {'class' : 'vehicle-card'})
    
    # loop through results
    for result in results:
    
        try:
            nama.append(result.find('h2').get_text()) 
        except:
            nama.append('n/a')
    
    # untuk mencari jarak tempuh
        try:
            jarak_tempuh.append(result.find('div', {'class':'mileage'}).get_text())
        except:
            jarak_tempuh.append('n/a')
    
    # untuk mencari nama dealer
        try:
            nama_dealer.append(result.find('div', {'class':'dealer-name'}).get_text().strip())
        except:
            nama_dealer.append('n/a')
    # untuk mencari rate
        try:
            rate.append(result.find('span', {'class':'sds-rating__count'}).get_text())
        except:
            rate.append('n/a')
    
    # untuk mencari banyak review
        try:
            banyak_review.append(result.find('span', {'class':'sds-rating__link'}).get_text())
        except:
            banyak_review.append('n/a')
        
    #untuk mencari harga 
        try:
            harga.append(result.find('span', {'class':'primary-price'}).get_text())
        except:
            harga.append('n/a')    
```

```python
dealer_mobil = pd.DataFrame({'nama': nama, 'jarak tempuh': jarak_tempuh, 'nama dealer': nama_dealer,'rate': rate, 'banyak review': banyak_review, 'harga': harga})
```
